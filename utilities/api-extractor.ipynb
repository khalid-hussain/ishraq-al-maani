```python
import requests

enTransAPI = "https://api.quran.com/api/v4/quran/translations/20?chapter_number=31"
arTextAPI = "https://api.quran.com/api/v4/quran/verses/uthmani?chapter_number=31"

enTransContent = requests.get(enTransAPI).json()
arTextContents = requests.get(arTextAPI).json()

listAR = []
listEN = []

for k in arTextContents['verses']:
  listAR.append(k['text_uthmani'])
  # print("{}".format(k['text_uthmani']))

for m in enTransContent['translations']:
  listEN.append(m['text'])
  # print("{}".format(m['text']))
```

```python
noAyaatAR = len(listAR)
noAyaatEN = len(listEN)
display(noAyaatAR)
display(noAyaatEN)
```

```python
for x in range(0,34):
  template = f"""
  \\begin{{edition*}}
  \\begin{{Arabic}}
  {listAR[x]}
  \end{{Arabic}}
  \end{{edition*}}

  \\begin{{translation}}
  {listEN[x]}
  \end{{translation}}
  """
  print(template)
```
